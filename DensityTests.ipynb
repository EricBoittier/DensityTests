```python
import os
import matplotlib.pyplot as plt
plt.style.use(['science','ieee', "no-latex"])
```

```python
files = [x for x in os.listdir() if x.__contains__(".text")]
errors = []
denominators = [1,2,4,8,16]
fractions = [1/x for x in denominators]

gridpoints = [(80*x)**3 for x in fractions]


for x in denominators:
    f = open(f"results_{x}.text").readlines()[21]
    result = float(f.split()[1])
    errors.append(result)
    
```

```python
plt.plot(gridpoints, errors, "x-")
plt.xlabel("$n$ Gridpoints")
plt.ylabel("ESP Error (kcal mol$^{-1}$ $e^{-1}$)")
plt.xscale("log")
```
